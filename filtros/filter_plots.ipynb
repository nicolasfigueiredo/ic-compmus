```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
y, sr = librosa.load("../audio/48054_44k.wav", sr=44100)
```

```python
y = y[44100*4:44100*8]
Y = librosa.stft(y, n_fft = 4096, hop_length = 512)
Y_dB = librosa.amplitude_to_db(Y, ref=np.max)
librosa.display.specshow(Y_dB, y_axis='linear', sr=sr, hop_length = 512)
```

```python
ys1, Ys1 = my_filters.my_filter(y,sr)
ys2, Ys2 = my_filters.my_filter2(y,sr)
ys3, Ys3 = my_filters.my_filter3(y,sr)
```

```python
plt.figure(figsize=(16,8))

plt.subplot(1,4,1)
librosa.display.specshow(Y_dB, y_axis='linear', sr=sr, hop_length = 512)
#plt.colorbar(format='%+2.0f dB')
plt.title('Original spectrogram')

plt.subplot(1,4,2)
librosa.display.specshow(librosa.amplitude_to_db(Ys1, np.max), y_axis='linear', sr=sr, hop_length = 512)
#plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram filtered using Filter A')

plt.subplot(1,4,3)
librosa.display.specshow(librosa.amplitude_to_db(Ys2, np.max), y_axis='linear', sr=sr, hop_length = 512)
#plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram filtered using Filter B')

plt.subplot(1,4,4)
librosa.display.specshow(librosa.amplitude_to_db(Ys3, np.max), y_axis='linear', sr=sr, hop_length = 512)
#plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram filtered using Filter C')
```
