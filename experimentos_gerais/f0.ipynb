```python
import librosa
import librosa.display
import numpy as np
import IPython.display
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
y, sr = librosa.load('../../db-tcc-felipe/Aegolius harrisii/song/100013.mp3.filtered.wav', sr=44100)
Y = librosa.stft(y)
Y_dB = librosa.amplitude_to_db(Y, ref=np.max)
```

```python
librosa.display.specshow(Y_dB, y_axis='log', sr=sr)
```

```python
pitches, magnitudes = librosa.core.piptrack(S=Y, sr=sr, threshold = 0.8)
```

```python
def detect_pitch(t):
    index = magnitudes[:, t].argmax()
    pitch = pitches[index, t]
    return pitch
```

```python
pitches_array = []
for i in range(0,magnitudes.shape[1]):
    pitches_array.append(detect_pitch(i))
```

```python
x = range(0, magnitudes.shape[1])
plt.bar(x, pitches_array)
```
