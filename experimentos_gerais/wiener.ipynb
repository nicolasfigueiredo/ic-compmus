```python
import librosa
import librosa.display
import scipy.signal
import numpy as np
import IPython.display
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
y, sr = librosa.load("pompeu.wav")
```

```python
Y = librosa.stft(y)
D = librosa.amplitude_to_db(Y, ref=np.max)
librosa.display.specshow(D, y_axis='linear')
plt.colorbar(format='%+2.0f dB')
plt.title('Linear-frequency power spectrogram')
```

```python
Y_wiener = scipy.signal.wiener(Y)
```

```python
D_wiener = librosa.amplitude_to_db(Y_wiener, ref=np.max)
librosa.display.specshow(D_wiener, y_axis='linear')
plt.colorbar(format='%+2.0f dB')
plt.title('Linear-frequency power spectrogram')
```

```python
y_wiener = librosa.core.istft(Y_wiener)
IPython.display.Audio(data=y_wiener, rate=sr)
```
