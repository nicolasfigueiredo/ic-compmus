```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
y, sr = librosa.load("../audio/tcc-f2/1000288.mp3", sr=44100)
Y = librosa.stft(y, n_fft = 4096, hop_length = 512)
Y_dB = librosa.amplitude_to_db(Y, ref=np.max)
```

```python
librosa.display.specshow(Y_dB, y_axis='linear', sr=sr, hop_length = 512)
#plt.colorbar(format='%+2.0f dB')
plt.title('Linear-frequency power spectrogram')
```

```python
mag, phase = librosa.magphase(Y)
plt.plot(mag.sum(axis=1))
```

```python
Y_achatado = mag.sum(axis=1)
```

```python
def CFAR(Y, num_cells_guard, num_cells_train, threshold):
    peaks = []
    num_cells_guard_half = round(num_cells_guard/2)
    num_cells_train_half = round(num_cells_train/2)
    num_cells_side = num_cells_guard_half + num_cells_train_half   # num celulas em cada lado da celula teste
    
    for i in range(num_cells_side, Y.size - num_cells_side):
        training_cells_1h = Y[i-num_cells_train_half-num_cells_guard_half:i-num_cells_guard_half]
        training_cells_2h = Y[i+1+num_cells_guard_half:i+1+num_cells_guard_half+num_cells_train_half]
        training_cells = np.concatenate([training_cells_1h, training_cells_2h])
        
        if Y[i] > np.mean(training_cells)*threshold:
            peaks.append(i)
              
#         if Y[i] > np.sort(training_cells)[-1]:
#             peaks.append(i)
    
    
    peaks = np.array(peaks, dtype=int)
    return peaks
```

```python
peaks = CFAR(Y_achatado, 8, 16, 1.5)
```

```python
peaks
```

```python
x = np.arange(Y_achatado.shape[0])
plt.plot(x, Y_achatado)
plt.plot(x[peaks], Y_achatado[peaks], 'rD')
plt.xlabel('x')
plt.ylabel('y')
```

```python
def CFAR_2D(Y, num_cells_guard, num_cells_train, threshold):
    peaks = CFAR(Y.sum(axis=1), num_cells_guard, num_cells_train, 1.5)
    new_peaks = []
    for peak in peaks:
        Y_partial = Y[peak,:]
        for new_peak in CFAR(Y_partial, num_cells_guard, num_cells_train, threshold):
            new_peaks.append([peak,new_peak])
        
    return new_peaks
```

```python
peaks = CFAR_2D(mag, 8, 16, 4)
peaks_toplot = np.array(peaks, dtype=int)
y, x = peaks_toplot.T
```

```python
librosa.display.specshow(Y_dB, y_axis='linear', sr=sr, hop_length = 512)
plt.scatter(x, y)
```
