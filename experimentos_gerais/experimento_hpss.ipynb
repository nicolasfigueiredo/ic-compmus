```python
import librosa
import librosa.display
import IPython.display
import numpy as np
```

```python
import matplotlib.pyplot as plt
import matplotlib.style as ms
ms.use('seaborn-muted')
%matplotlib inline
```

```python
y, sr = librosa.load("pompeu.wav")
```

```python
IPython.display.Audio(data=y, rate=sr)
```

```python
y_h, y_p = librosa.effects.hpss(y)
```

```python
IPython.display.Audio(data=y_h, rate=sr)
```

```python
IPython.display.Audio(data=y_p, rate=sr)
```

```python
y_h2, y_p2 = librosa.effects.hpss(y_h)
```

```python
IPython.display.Audio(data=y_h2, rate=sr)
```

```python
spectrum = librosa.core.stft(y)
```

```python
H, P = librosa.decompose.hpss(spectrum, margin=1.5)

R = spectrum - (H+P)
y_harm = librosa.core.istft(H)
y_perc = librosa.core.istft(P)
y_resi = librosa.core.istft(R)
```

```python
IPython.display.Audio(data=y_harm, rate=sr)
```

```python
IPython.display.Audio(data=y_perc, rate=sr)
```

```python
IPython.display.Audio(data=y_resi, rate=sr)
```
